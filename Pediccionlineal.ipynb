```python
import pandas as pd
datos = pd.read_excel("/content/drive/MyDrive/carpeta estadistica/c3/datos/datos_industrias.xlsx")
datos.head()
```

```python
# prompt: necesito un analisis de regresion lineal simple que la variable x sea "Empleados" sea y la variable y sea "Tasa_Crecimiento" de nuestra base de datos. dame las conclusiones. y graficalo y dame la ecuacion de la recta

import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming 'datos' is already loaded from the previous code

# Define independent and dependent variables
X = datos['Empleados']
y = datos['Tasa_Crecimiento']

# Add a constant to the independent variable for the intercept
X = sm.add_constant(X)

# Fit the regression model
model = sm.OLS(y, X).fit()

# Print the summary of the regression
print(model.summary())

# Get the equation of the line
intercept = model.params['const']
slope = model.params['Empleados']
print(f"\nEcuación de la recta: Tasa_Crecimiento = {intercept:.4f} + {slope:.4f} * Empleados")

# Plotting the regression line
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Empleados', y='Tasa_Crecimiento', data=datos)
plt.plot(datos['Empleados'], model.predict(sm.add_constant(datos['Empleados'])), color='red', linewidth=2)
plt.title('Regresión Lineal Simple: Tasa de Crecimiento vs Número de Empleados')
plt.xlabel('Número de Empleados')
plt.ylabel('Tasa de Crecimiento')
plt.grid(True)
plt.show()

# Conclusiones
print("\nConclusiones del análisis de regresión lineal simple:")
print(f"- La ecuación de la recta de regresión es: Tasa_Crecimiento = {intercept:.4f} + {slope:.4f} * Empleados")
print(f"- El coeficiente de la variable 'Empleados' es {slope:.4f}. Esto indica que por cada empleado adicional, la Tasa de Crecimiento se espera que aumente en {slope:.4f} unidades (en promedio), manteniendo todo lo demás constante.")
print(f"- El valor p para la variable 'Empleados' es {model.pvalues['Empleados']:.4f}.")
if model.pvalues['Empleados'] < 0.05:
    print("  Como el valor p es menor que 0.05, podemos concluir que el número de empleados tiene un efecto estadísticamente significativo en la Tasa de Crecimiento.")
else:
    print("  Como el valor p es mayor que 0.05, no podemos concluir que el número de empleados tenga un efecto estadísticamente significativo en la Tasa de Crecimiento al nivel de significancia del 5%.")
print(f"- El R-cuadrado ajustado es {model.rsquared_adj:.4f}. Esto significa que el {model.rsquared_adj*100:.2f}% de la variabilidad en la Tasa de Crecimiento es explicado por el número de empleados.")
print(f"- El valor p del test F es {model.f_pvalue:.4f}.")
if model.f_pvalue < 0.05:
     print("  Como el valor p del test F es menor que 0.05, el modelo de regresión en su conjunto es estadísticamente significativo, lo que sugiere que al menos uno de los predictores (en este caso, 'Empleados') explica una proporción significativa de la varianza en la variable dependiente.")
else:
     print("  Como el valor p del test F es mayor que 0.05, el modelo de regresión en su conjunto no es estadísticamente significativo.")
```

```python
# prompt: necesito un analisis de regresion lineal simple que la variable x sea "Salario" sea y la variable y sea "Tasa_Crecimiento" de nuestra base de datos. dame las conclusiones. y graficalo y dame la ecuacion de la recta

import matplotlib.pyplot as plt
import statsmodels.api as sm

# Ajuste del modelo de regresión lineal
X = datos['Salario']
y = datos['Tasa_Crecimiento']
X = sm.add_constant(X)  # Agrega una constante para el intercepto
modelo = sm.OLS(y, X).fit()

# Resultados del modelo
print(modelo.summary())

# Gráfico de dispersión con la línea de regresión
plt.figure(figsize=(8, 6))
plt.scatter(datos['Salario'], datos['Tasa_Crecimiento'], label='Datos')
plt.plot(datos['Salario'], modelo.fittedvalues, color='red', label='Regresión Lineal')
plt.xlabel('Salario')
plt.ylabel('Tasa de Crecimiento')
plt.title('Regresión Lineal Simple')
plt.legend()
plt.show()

# Ecuación de la recta
pendiente = modelo.params['Salario']
intercepto = modelo.params['const']
print(f"Ecuación de la recta: y = {pendiente:.2f}x + {intercepto:.2f}")

# Conclusiones (ejemplo, adaptar según los resultados del modelo)
print("\nConclusiones:")
if modelo.pvalues['Salario'] < 0.05:
    print("Existe una relación estadísticamente significativa entre el salario y la tasa de crecimiento.")
    if pendiente > 0:
        print("A medida que aumenta el salario, la tasa de crecimiento tiende a aumentar.")
    else:
        print("A medida que aumenta el salario, la tasa de crecimiento tiende a disminuir.")
else:
    print("No existe una relación estadísticamente significativa entre el salario y la tasa de crecimiento.")

r_cuadrado = modelo.rsquared
print(f"R-cuadrado: {r_cuadrado:.2f}")
print("El R-cuadrado indica la proporción de la variabilidad de la tasa de crecimiento que es explicada por el salario.")
```
